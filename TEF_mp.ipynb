#**Ajuste de m√≠nimos quadrados utilizando o Python.**


```python
import numpy as np
import pandas as pd
import scipy as sp
import astropy as ap
import matplotlib.pyplot as plt


x_coord = np.arange(0, 100, 2)
y_coord = (1e0 + 2e0 * x_coord) + (np.random.randn(50) * 1e1)


# Definindo os elementos em uma matriz a partir do sistema linear:
Matriz_11 = np.sum(x_coord ** 2)
Matriz_12 = np.sum(x_coord)
Matriz_21 = Matriz_12
Matriz_22 = len(x_coord)
Coluna_11 = np.sum(x_coord * y_coord)
Colina_21 = np.sum(y_coord)

# Escrevendo a matriz:
Matriz = np.array([[Matriz_11, Matriz_12], [Matriz_21, Matriz_22]])
Matriz_coluna = np.array([Coluna_11, Colina_21])

# Resolvendo o sistema linear:
S_linear = np.linalg.solve(Matriz, Matriz_coluna)
S_linear

yFit =(S_linear[0] * x_coord) + S_linear[1]


plt.plot(x_coord, y_coord, '*')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Ajuste de m√≠nimos quadrados")
plt.plot(x_coord, yFit)
plt.show()
```
